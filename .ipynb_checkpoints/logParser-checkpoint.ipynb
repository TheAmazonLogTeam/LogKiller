```python
def getLog(fileName):
    """
    Ouvre le fichier et cr√®e une liste avec un log par ligne
    """
    results = []
    with open(fileName, newline='') as inputfile:
        for row in csv.reader(inputfile):
            results.append(row)
    return results
```

```python
# Liste des fichiers logs
glob.glob('/Users/maxime/Desktop/logs/*.log')
```

```python
# Utilise le module Glob pour Lister tous les fichiers logs et les ouvrir
res = []
for filename in glob.glob('logs/*.log'):
    res.append(getLog(filename))

# Flatten ma liste --> passer d'une "liste de liste" √† une liste simple
logs = list(chain(*res))

# Split les log en fonction des 3 premiersd espaces rencontr√©s
logSplit = list(map(lambda x : logs[x][0].split(' ', 3),
                    range(len(logs))))

# Cr√©ation d'un DataFrame √† la place de la liste Split√©e
df = pd.DataFrame(logSplit, columns=["Date", "Hostname", "Command", "message"])

# Cleaning
df["Heure"] = df["Date"].str.split("T",1).str[1].str[:-6]
df["Date"] = df["Date"].str.split("T",1).str[0]
df["Command"] = df["Command"].str[:-1]
df["ID"] = df["Command"].str.split("[").str[1].str[:-1]
df["Command"] = df["Command"].str.split("[").str[0]
df = df[["Date", "Heure", "Hostname", "Command", "ID", "message"]]

# Index
df = df.set_index("Date")

df.head(20)
```

## TF-IDF

```python
from sklearn.feature_extraction.text import *
import numpy as np
```

```python
def get_scores(vectorizer, tfidf_result):
    scores = zip(vectorizer.get_feature_names(),
                 np.asarray(tfidf_result.sum(axis=0)).ravel())
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    
    scores_df = pd.DataFrame(data=sorted_scores,
                 columns=['word','tf-idf score'])
    
    return scores_df
```

```python
message = df['message'].ravel()
vectorizer = TfidfVectorizer(token_pattern="(\S*)")
tfidf_result = vectorizer.fit_transform(message)
scores_df = get_scores(vectorizer, tfidf_result)
```
